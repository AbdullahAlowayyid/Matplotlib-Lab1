```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
Movies = ['Interstellar', 'The Godfather ', 'The Dark Knight', 'The Shawshank Redemption']
Rating  = [8.7, 9.2, 9.1, 9.3]
plt.plot(Movies, Rating, color = 'b')
plt.title('Top 5 Movies Rating Chart 1',fontsize = 15)
plt.xlabel('Movies',fontsize = 15)
plt.ylabel('Rating',fontsize = 15)
plt.show()

#####################################
data = {
    'Movies' : ['Interstellar', 'The Godfather ', 'The Dark Knight', 'The Shawshank Redemption'],
    'Ratings' : [8.7, 9.2, 9.1, 9.3]
}
df = pd.DataFrame(data, columns = ['Movies', 'Ratings'])

plt.bar(df['Movies'], df['Ratings'], color = 'b')
plt.title('Top 5 Movies Rating Chart 2',fontsize = 15)
plt.xlabel('Movies',fontsize = 15)
plt.ylabel('Rating',fontsize = 15)
plt.yticks(range(1, 10));
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
games_genre = ['AAA', 'MMO', 'MOBA', 'FPS', 'SPORTS']
pop = [12789453, 30562321, 19537583, 39903202, 8908332]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#ffcc39']

plt.title('Video Games Genres Popularity', fontsize = 15)
plt.pie(pop, labels = games_genre, autopct = '%1.1f%%', colors = colors,
        startangle = 180, radius = 1, explode = [0.1, 0, 0.2, 0, 0])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Movies'], df['Ratings'], marker = 'o', alpha = 0.4, s = 50)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['Movies'], df['Ratings'], marker = 'o', alpha = 0.4, s = 50)
plt.title('Top 5 Movies Rating', loc = 'left', color = 'red')
plt.xlabel('Movies', fontsize = 16)
plt.ylabel('Ratings', fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
years = [1960, 1970, 1980, 1990, 2000]

plt.plot(years, [12,21,15,34,60], color = 'g')
plt.plot(years, [40,50,20,30,10], color ='b')
plt.plot(years, [20, 20, 5, 10, 40], color ='y')

plt.title('Users Track of 3 Apps')
plt.legend(['App 1', 'App 2', 'App 3'],loc = [0.01, 0.4])
plt.xlabel('Years')
plt.ylabel('Users in millions');
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize = [20,10])
plt.suptitle('Top 5 Movies Rating',fontsize = 20)

Movies = ['Interstellar', 'The Godfather ', 'The Dark Knight', 'The Shawshank Redemption']
Rating  = [8.7, 9.2, 9.1, 9.3]

plt.subplot(1,3,1)
plt.plot(Movies, Rating, color = 'b')
plt.title('Chart 1',fontsize = 15)
plt.xlabel('Movies',fontsize = 15)
plt.ylabel('Rating',fontsize = 15)

print("\n")

plt.subplot(1,3,2)
plt.bar(Movies, Rating, color = 'b')
plt.title('Chart 2',fontsize = 15)
plt.xlabel('Movies',fontsize = 15)
plt.ylabel('Rating',fontsize = 15)
plt.yticks(range(1, 10));

print("\n")

plt.subplot(1, 3, 3)
plt.scatter(Movies, Rating, color = 'b', alpha = 0.3)
plt.title('Chart 3',fontsize = 15)
plt.xlabel('Movies',fontsize = 15)
plt.ylabel('Rating',fontsize = 15);
```
